```python
coin_possibilities = ['H', 'T']

for coin1 in coin_possibilities:
    print(coin1)
```

```python
for coin1 in coin_possibilities:
    for coin2 in coin_possibilities:
        print(coin1, coin2)
```

```python
sum = 0
for coin1 in coin_possibilities:
    for coin2 in coin_possibilities:
        for coin3 in coin_possibilities:
            for coin4 in coin_possibilities:
                sum += 1
                print(coin1, coin2, coin3, coin4)

print(f'{sum} possibilities')
```

```python
sum = 0
has_1_head = 0
has_5_heads = 0
for coin1 in coin_possibilities:
    for coin2 in coin_possibilities:
        for coin3 in coin_possibilities:
            for coin4 in coin_possibilities:
                for coin5 in coin_possibilities:
                    for coin6 in coin_possibilities:
                        for coin7 in coin_possibilities:
                            for coin8 in coin_possibilities:
                                for coin9 in coin_possibilities:
                                    for coin10 in coin_possibilities:
                                        sum += 1
                                        number_of_heads = (coin1 == 'H') + (coin2 == 'H') + (coin3 == 'H') + (coin4 == 'H') + (coin5 == 'H') + (coin6 == 'H') + (coin7 == 'H') + (coin8 == 'H') + (coin9 == 'H') + (coin10 == 'H')
                                        if number_of_heads == 1: has_1_head += 1
                                        if number_of_heads == 5: has_5_heads += 1
                                        print(coin1, coin2, coin3, coin4, coin5, coin6, coin7, coin8, coin9, coin10, number_of_heads)

print(f'{sum} possibilities')
print(f'{has_1_head} had 1 head and {has_5_heads} had 5 heads')
```
